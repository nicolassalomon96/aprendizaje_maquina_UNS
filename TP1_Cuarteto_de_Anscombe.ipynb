## Consigna - Trabajo Práctico Nº1
Dado 4 datasets con pares (x,y), ¿a que dataset corresponden estos nuevos pares de datos? ([16,8];[8,6];[16:10] y [16,6]). Aclaración: cada uno corresponde a un único dataset

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
D1 = [(10,8.04),(8,6.95),(13,7.58),(9,8.81),(11,8.33),(14,9.96),(6,7.24),(4,4.26),(12,10.84),(7,4.82),(5,5.68)]
D2 = [(10,9.14),(8,8.14),(13,8.74),(9,8.77),(11,9.26),(14,8.1),(6,6.13),(4,3.1),(12,9.13),(7,7.26),(5,4.74)]
D3 = [(10,7.46),(8,6.77),(13,12.74),(9,7.11),(11,7.81),(14,8.84),(6,6.08),(4,5.39),(12,8.15),(7,6.42),(5,5.73)]
D4 = [(8,6.58),(8,5.76),(8,7.71),(8,8.84),(8,8.47),(8,7.04),(8,5.25),(19,12.5),(8,5.56),(8,7.91),(8,6.89)]
```

```python
#Separo la lista y tuplas para graficar
x1 = []; x2 = []; x3 = []; x4 = []
y1 = []; y2 = []; y3 = []; y4 = []
[x1.append(i[0]) for i in D1]
[x2.append(i[0]) for i in D2]
[x3.append(i[0]) for i in D3]
[x4.append(i[0]) for i in D4]
[y1.append(i[1]) for i in D1]
[y2.append(i[1]) for i in D2]
[y3.append(i[1]) for i in D3]
[y4.append(i[1]) for i in D4]

#Convierto los arrays en arreglos numpy para facilitar su manejo y gráfica
x1 = np.asarray(x1)
x2 = np.asarray(x2)
x3 = np.asarray(x3)
x4 = np.asarray(x4)
y1 = np.asarray(y1)
y2 = np.asarray(y2)
y3 = np.asarray(y3)
y4 = np.asarray(y4)

#Me aseguro que se grabaron correctamente, mostrando por pantalla dos arreglos
print(x1)
print(y3)
```

```python
#Grafico cada dataset para tener una representación gráfica de cada valor

fig, axs = plt.subplots(2, 2)
fig.set_size_inches(10,10)
fig.suptitle('Gráfica de dispersión', fontsize=16, y=1.05)

axs[0,0].scatter(x1,y1,marker='o')
axs[0,0].set_title('Dataset 1')
axs[0,0].set_xlabel('Valores de x')
axs[0,0].set_ylabel('Valores de y')
axs[0,0].set_xlim([0, 18])
axs[0,0].set_ylim([0, 12])

axs[0,1].scatter(x2,y2,marker='o')
axs[0,1].set_title('Dataset 2')
axs[0,1].set_xlabel('Valores de x')
axs[0,1].set_ylabel('Valores de y')
axs[0,1].set_xlim([0, 18])
axs[0,1].set_ylim([0, 12])

axs[1,0].scatter(x3,y3,marker='o')
axs[1,0].set_title('Dataset 3')
axs[1,0].set_xlabel('Valores de x')
axs[1,0].set_ylabel('Valores de y')
axs[1,0].set_xlim([0, 18])
axs[1,0].set_ylim([0, 12])

axs[1,1].scatter(x4,y4,marker='o')
axs[1,1].set_title('Dataset 4')
axs[1,1].set_xlabel('Valores de x')
axs[1,1].set_ylabel('Valores de y')
axs[1,1].set_xlim([0, 20])
axs[1,1].set_ylim([0, 13])

fig.tight_layout()
plt.show()
```

Luego de analizar los gráficos, podemos decir que:

-El punto [16,8] corresponde al dataset 1

-El punto [8,6] corresponde al dataset 4

-El punto [16,10] corresponde al dataset 3

-El punto [16,6] corresponde al dataset 2

```python
#Grafico cada dataset para tener una representación gráfica de cada valor, agregando los puntos mencionados anteriormente
#(en color rojo)

fig, axs = plt.subplots(2, 2)
fig.set_size_inches(10,10)
fig.suptitle('Gráfica de dispersión', fontsize=16, y=1.05)

axs[0,0].scatter(x1,y1,marker='o')
axs[0,0].set_title('Dataset 1')
axs[0,0].set_xlabel('Valores de x')
axs[0,0].set_ylabel('Valores de y')
axs[0,0].set_xlim([0, 18])
axs[0,0].set_ylim([0, 12])
axs[0,0].scatter(16,8,marker='o',color='red')

axs[0,1].scatter(x2,y2,marker='o')
axs[0,1].set_title('Dataset 2')
axs[0,1].set_xlabel('Valores de x')
axs[0,1].set_ylabel('Valores de y')
axs[0,1].set_xlim([0, 18])
axs[0,1].set_ylim([0, 12])
axs[0,1].scatter(16,6,marker='o',color='red')

axs[1,0].scatter(x3,y3,marker='o')
axs[1,0].set_title('Dataset 3')
axs[1,0].set_xlabel('Valores de x')
axs[1,0].set_ylabel('Valores de y')
axs[1,0].set_xlim([0, 18])
axs[1,0].set_ylim([0, 12])
axs[1,0].scatter(16,10,marker='o',color='red')

axs[1,1].scatter(x4,y4,marker='o')
axs[1,1].set_title('Dataset 4')
axs[1,1].set_xlabel('Valores de x')
axs[1,1].set_ylabel('Valores de y')
axs[1,1].set_xlim([0, 20])
axs[1,1].set_ylim([0, 13])
axs[1,1].scatter(8,6,marker='o',color='red')

fig.tight_layout()
plt.show()
```

## Conclusión:
Este Trabajo Práctico nos muestra la importancia que tiene el representar graficamente cada uno de los valores que tengamos en nuestro dataset antes de tomar alguna decisión. Ya que puede suceder, como este caso, que los datasets tengan los mismos parámetros estadísticos (medias, varianzas, correlación, línea de regreción), pero al representar los valores graficamente nos damos cuenta a simple vista, a cual dataset pertenece cada nuevo punto (par de valores [x,y]).
